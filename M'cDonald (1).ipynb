```python
df_cleaned.to_csv('Nutrical Dataset.csv', index=False)
#EXPLORATORY DATA ANALYSIS

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize = (10,6))
sns.histplot(df['Calories'],bins =20,kde =True,color = 'skyblue')
plt.title('Distribution of calorie counts cross menu items',fontsize =16)
plt.xlabel('calories',fontsize=14)
plt.ylabel('Frequency',fontsize=14)
plt.show()

plt.figure(figsize=(10,6))
sns.boxplot(x=df['Calories'], color='lightgreen')
plt.title('Box Plot of Calorie Counts Across Menu Items', fontsize=16)
plt.xlabel('Calories', fontsize=14)
plt.show()


plt.figure(figsize=(10, 6))
sns.kdeplot(df['Calories'], shade=True, color='red')
plt.title('Density Plot of Calorie Counts Across Menu Items', fontsize=16)
plt.xlabel('Calories', fontsize=14)
plt.ylabel('Density', fontsize=14)
plt.show()










```

```python
print(df.columns)
print(df[['Item','Calories','Total Fat','Protein','Carbohydrates']].head())
nutrients = df[['Total Fat','Protein','Carbohydrates']]
print(nutrients.describe())

sns.pairplot(df[['Total Fat', 'Protein', 'Carbohydrates']], diag_kind='kde', corner=True)
plt.suptitle('Pair Plot of Nutritional Content', y=1.02, fontsize=16)
plt.show()



corr = df[['Total Fat','Protein','Carbohydrates']].corr()
plt.figure(figsize =(8,6))
sns.heatmap(corr,annot = True)
plt.title('Correlation Matrix of Nutritional Content',fontsize=16)
plt.show()
```

```python
print(df.columns)
top_items = df[['Item','Calories','Total Fat','Protein','Carbohydrates']].nlargest(10,'Calories')
top_items_melted = top_items.melt(id_vars='Item', value_vars=['Calories', 'Total Fat', 'Protein', 'Carbohydrates'],
                                  var_name='Nutrient', value_name='Amount')

plt.figure(figsize=(12, 6))
sns.barplot(x='Item', y='Amount', hue='Nutrient', data=top_items_melted)
plt.xticks(rotation=45, ha='right')
plt.title('Nutritional Content of Top 10 Calorie Items', fontsize=16)
plt.show()
```
